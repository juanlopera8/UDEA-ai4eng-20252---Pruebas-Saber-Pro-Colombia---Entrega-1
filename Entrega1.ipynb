```python
# 01_exploracion.ipynb - Código mínimo
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1) Cargar dataset
# Asegúrate de tener train.csv en la misma carpeta que el notebook
train = pd.read_csv("train.csv")
print("Dimensiones del dataset:", train.shape)

# 2) Primera vista
display(train.head())

# 3) Distribución de la variable objetivo
if "RENDIMIENTO_GLOBAL" in train.columns:
    print("\nDistribución de RENDIMIENTO_GLOBAL:")
    print(train["RENDIMIENTO_GLOBAL"].value_counts())
    plt.figure(figsize=(6,4))
    train["RENDIMIENTO_GLOBAL"].value_counts().plot(kind="bar")
    plt.title("Distribución de RENDIMIENTO_GLOBAL")
    plt.xlabel("Clase")
    plt.ylabel("Número de estudiantes")
    plt.show()
else:
    print("La columna 'RENDIMIENTO_GLOBAL' no está presente.")

# 4) Ejemplo: distribución por 'ESTRATO' si existe
if "ESTRATO" in train.columns:
    plt.figure(figsize=(6,4))
    sns.countplot(data=train, x="ESTRATO", order=train["ESTRATO"].value_counts().index)
    plt.title("Distribución por ESTRATO")
    plt.show()

# 5) Valores faltantes
print("\nTop 10 columnas con más valores faltantes (porcentaje):")
print((train.isnull().mean()*100).sort_values(ascending=False).head(10))
```
